```python
# TO FILL THE NULL VALUES WITH APPROPIATE VALUE USING AGGREGATE FUNCTION 
```

```python
def impute_median(series):
    return series.fillna(series .median())
```

```python
review_data.Rating = review_data['Rating'].transform(impute_median)
```

```python
#TO CONVERT PRICE, REVIEW AND RATING INTO NUMERICAL VAULES
```

```python
review_data['Price'] = review_data['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
review_data['Price'] = review_data['Price'].apply(lambda x: float(x))
review_data['Reviews'] = pd.to_numeric(review_data['Reviews'], errors='coerce')
```

```python
review_data['Installs'] = review_data['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))
review_data['Installs'] = review_data['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))
review_data['Installs'] = review_data['Installs'].apply(lambda x: float(x))
```

```python
#TO PERFORM DATA VISUALIZATION ON THE DATA
```

```python
grp = review_data.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
#TO CREATE A PLOT FOR DATA VISUALIZATION OF THE DATA
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, 'bo')
plt.xticks(rotation=90)
plt.show()
```
